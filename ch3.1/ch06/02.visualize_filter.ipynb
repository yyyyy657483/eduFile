```python
import numpy as np
import matplotlib.pyplot as plt
from ch06.simple_convnet import SimpleConvNet

def filter_show(filters, nx=8, margin=3, scale=10):
    """
    c.f. https://gist.github.com/aidiary/07d530d5e08011832b12#file-draw_weight-py
    """
    FN, C, FM, FW = filters.shape
    ny = int(np.ceil(FN/nx))

    fig = plt.figure()
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.005, wspace=0.05)

    for i in range(FN):
        ax = fig.add_subplot(ny, nx, i+1, xticks=[], yticks=[])
        ax.imshow(filters[i, 0], cmap=plt.cm.gray_r, interpolation='nearest')
    
    plt.show()

# 학습 전 무작위(랜덤)의 가중치 생성된 filter 표시
network = SimpleConvNet()
filter_show(network.params['W1'])
print("-"*100)
# 학습된 가중치 표시
network.load_params("ch06/params.pkl")
filter_show(network.params['W1'])
```
